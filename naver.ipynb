```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

import csv

base_url = 'https://finance.naver.com/item/board.nhn?code='
```

```python
# crwaling module
class naver_crawler:
    
    def parse(self,text):
        
        # fetch post list
        soup = BeautifulSoup(text,'html.parser')
        
        find = soup.select('td.title > a')
        print(find)
        # all href in post list
        hrefs = list()
        for i in range(0,20):
            if(find):
                hrefs.append(str(find[i])[9:86])
        # fetch data

        for href in hrefs:
            a = href.replace("amp;","")
            response = requests.get('http://finance.naver.com'+a)
            soup = BeautifulSoup(response.text,'lxml')
            title = soup.find('strong',{'class':'c p15'})
            content = soup.find('div',{'id':'body','class':'view_se'})
            views = soup.find('span',{'class':'tah p11'})
            like = soup.find('strong',{'class':'tah p11 red01 _goodCnt'})
            unlike = soup.find('strong',{'class':'tah p11 blue01 _badCnt'})
            content = soup.find('div',{'id':'body','class':'view_se'})
            try:
                ip = soup.find('span',{'class':'gray03'})
                [s.extract() for s in ip]
                ip = ip.text
            except:
                print('fail')
            date = soup.find('th',{'class':'gray03 p9 tah'})
            if(title and content and views and like and unlike ):
                # return row data to write csv file
                self.data.append([date.string,views.string,title.string,content.get_text(separator='\n'),like.string,unlike.string,ip])

    # firts, get html
    def __init__(self,code):
        
        # data header
        self.columns = ['Date','Views','Title','Content','Like','Unlike','Ip']
        self.data = list()
        

        # crawling all pages
        i=1
        while(i<3000):
            page ='&page={}'.format(i)
            headers = {'Content-Type': 'application/json; charset=utf-8'}
            print(requests.get("http://naver.com",headers=headers))
            response = requests.get(base_url+code+page)
            if(response.status_code == 200):
                print(response.text)
                self.parse(response.text) 

            else:
                # if response status is not found
                print('end')
            if(i%100==0):
                print(i/3000+"%")
                pd.DataFrame(self.data,columns=self.columns).to_csv(code+'.csv',mode='w')
            i+=1
        pd.DataFrame(self.data,columns=self.columns).to_csv(code+'.csv',mode='w')
        return print('done')
```
